```python
df = pd.read_csv('Act3/antropometria-dataset-2.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          values = ['peso'],
                          index = ['desc_ent'],
                          aggfunc = {'peso':np.mean})
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          values = ['peso'],
                          index = ['desc_ent'],
                          columns = ['sexo'],
                          aggfunc = ['median']
                                    )
my_pivot
```

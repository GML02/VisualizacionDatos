## Pregunta 1: ¿Existe alguna correlación entre las ventas de Decoración y Papelería?

```python
sns.lmplot(x = 'Decoración',
           y = 'Papelería',
           data = df,
           fit_reg = False)
plt.title('Decoración vs Papelería', fontsize = 15)
```

Podemos ver que no existe una correlación entre las ventas de Decoración y Papelería, por lo que se puede asumir que no todo lo que se compra en la papelería es utilizado para fines decorativos y viceversa. 

## Pregunta 2: ¿Cuál es la relación entre la venta de Comidas preparadas y Jugos?


```python
sns.scatterplot(x = 'Comidas \npreparadas', y = 'Jugos', data = df)
sns.set(rc={'figure.figsize':(8,8)})
plt.title('¿Cuantas comidas preparadas y jugos se compran al mismo tiempo?')
plt.show()
```

Se puede observar en la gráfica anterior que el comportamiento de ventas de estos dos departamentos no comparten mucha relación tampoco, y que las comidas preparadas varían en un rango de 3 a 30 por día mientras que la venta de unidades de jugo varía entre las 25 y 55 unidades diarias. 

## Pregunta 3 : ¿Con que frecuencia se compran juguetes en todas las tiendas?

```python
df['Juguetería'].plot(kind = 'hist',
              figsize =(10,10))
```

En el histograma anterior podemos ver la frecuencia de unidades vendidas del departamento de Juguetería. Viendo la ilustración, podemos observar que es muy frecuente que se vendan entre 60 y 70 unidades de juguetes en todas las tiendas en general mientras que lo menos frecuente varía entre 20 y 30 unidades, y entre 90 y 100 unidades. 

## Pregunta 4: ¿Cuál es la tienda que más asadores ha vendido? 

```python
pivot_A = pd.pivot_table(df,index= ['Tienda'],values = ['Asadores'])
pivot_A = pivot_A.sort_values('Asadores', axis=0,ascending=False)
pivot_A[:]
```

```python
datos = pivot_A[:]
datos.plot(kind='bar')
```

Lo que se hizo para encontrar este dato fue primero desplegar los datos del departamento de Asadores en orden de mayor a menor y después de eso graficar esos datos con un gráfico de barras. Se puede observar que la tienda con mayor ventas de unidades de Asadores es la Tienda 7 con 47.25 unidades, mientras que la Tienda 5 solo vendió 10.95 unidades.

## Pregunta 7: ¿Hay relación entre las ventas de Frutas y Verduras?

```python
model = sm.OLS.from_formula('Frutas ~ Verduras', data = df)
result = model.fit()
result.summary()
```

En este análisis de regresión, se tomaron en cuenta solo las variables de Frutas y Verduras para ver la relación entre ellas pero podemos ver que el valor de R-cuadrado es muy bajo, por lo que concluimos que el ajuste del modelo no es bueno. Además, el valor de condición es muy alto, por lo que probablemente haya multicolinealidad entre los datos, como bien indica el programa. 

## Pregunta 8: ¿Cómo serian los porcentajes promedio de ventas de los primeros 5 departamentos?

```python
labels = "Frutas","Verduras","Lácteos","Charcutería","Enlatados"
sizes = [50.86,124.5,131.3,117.86,59.4]
explode = (0.1,0.2,0.2,0.2,0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
fig1.show()                                
    
```

Este diagrama de pie nos indica que porcentaje de ventas tiene cada uno de estos departamentos en todas las tiendas, tomando en cuenta sus valores promedio de ventas calculados en el archivo de Excel.

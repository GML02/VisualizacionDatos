# Pregunta de investigación 2: 

## ¿Como es el comportamiento de las ventas de "Frutas" y "Verduras" en las primeras 5 tiendas? 

```python
data = {'Ventas': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4', 'Tienda 5'],
        'Frutas': [30.990, 77.190, 65.775, 70.065, 51.840],
        'Verduras': [121.20, 120.90, 128.70, 122.25, 126.60],
       }
df = pd.DataFrame(data)
df
```

Aquí, podemos observar las ventas diarias de las primeras 5 tiendas en los departamentos de Frutas y Vrduras. Con esta pequeña tabla, podemos comparar en una gráfica de lineas el comportamiento de las ventas de ambos departamentos. 

```python
df.plot(kind = 'line')
plt.xlabel('Tiendas', fontsize = 16)
plt.ylabel('Ventas', fontsize = 16)

#Ventas de Frutas
plt.plot(0,30.99,'x')
plt.plot(1,77.190,'x')
plt.plot(2,65.775,'x')
plt.plot(3,70.065,'x')
plt.plot(4,51.840,'x')

#Ventas de Verduras
plt.plot(0,121.20,'o')
plt.plot(1,120.90,'o')
plt.plot(2,128.70,'o')
plt.plot(3,122.25,'o')
plt.plot(4,126.60,'o')
```

El gráfico superior de línea nos muestra el comportamiento de las ventas de cada departamento (Frutas y Verduras) y los puntos específicos de la cantidad de unidades vendidas. Como se ve en la gráfica, las ventas de "Verduras" son mucho mayores a las ventas de "Frutas". 

# Pregunta de investigación 3:

## ¿Hay alguna correlación entre las ventas de "Deportes" y "Zapatería"? ¿Cómo son las ventas de cada departamento?

```python
df = pd.read_csv('SituProb_VD.csv') # Para que el programa corra, hay que volver a definir nuestra base de datos como df. 
```

```python
df.loc[0:29,'Deportes':'Zapatería']
```

```python
data2 = {'Ventas': ['Tienda 30', 'Tienda 29', 'Tienda 28', 'Tienda 27', 'Tienda 26'],
        'Deportes': [72.75, 64.05, 70.20, 60.00, 92.10],
        'Zapatería': [89.318, 31.248, 50.550, 50.121, 44.262],
       }
df2 = pd.DataFrame(data2)
df2
```

```python
sns.lmplot (x = 'Deportes',
            y = 'Zapatería',
            data = df2,
            fit_reg = False,
           )
plt.title('Scatterplot of Deportes v Zapatería', fontsize = 16)
plt.xticks(fontsize = 15)
plt.show()
```

En el análisis de correlación entre variables, no se encontró ninguna correlación entre las ventas de los departamentos de "Deportes" y "Zapatería" tomando en cuenta los datos de las últimas 5 tiendas. 

```python
sns.boxplot(data = df.loc[:, ['Deportes', 'Zapatería']],
            )
```

Utilizando nuevamente una gráfica de caja, podemos ver como se comportan los datos de ambos departamentos. En el de "Deportes", la mayor parte de los datos se concentra en un rango de entre 60 y 75 unidades vendidas, con un dato atípico por encima de las 90 unidades. 

Por otro lado, en el departamento de "Zapatería", la mayor concentración de ventas se encuentra en un rango entre 40 y 50 ventas, mientras que tiene 2 datos atípicos: uno alrededor de las 30 ventas y otro entre 80 y 90 ventas.

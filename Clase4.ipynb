## Pregunta de investigacion: 1
### Existe alguna correlacion entre 'Wingspan', y 'CWDistance'?

```python
sns.lmplot (x = 'Wingspan',
            y = 'CWDistance',
            data = df,
            fit_reg = False,
            hue = 'Gender'
           )
plt.title('Scatterplot of Wingspan vs CWDistance', fontsize = 17)
plt.xticks(fontsize = 15)
plt.show()
```

## Conclusion 
No se encuentra ninguna correlacion entre las dos variables
